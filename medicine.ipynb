```python
def what_use(med_name):
    if med_name in df['name'].unique():
        return(df[df['name'] == med_name]['use0'].unique())
    else:
        return("Not found")
    
def side_effects(med_name):
    if med_name in df['name'].unique():
        return(df[df['name'] == med_name]['sideEffects0'].unique(), df[df['name'] == med_name]['sideEffects1'].unique(), 
                df[df['name'] == med_name]['sideEffects2'].unique())
    else:
        return("Not found")
    
def substitue_meds(med_name):
    if med_name in df['name'].unique():
        return(df[df['name'] == med_name]['substitute0'].unique(), df[df['name'] == med_name]['substitute1'].unique(), 
                df[df['name'] == med_name]['substitute2'].unique())
    else:
        return("Not found")
```
